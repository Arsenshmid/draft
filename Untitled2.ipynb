график числа полностью вакцинированных людей в Китае по времени.

```python
import seaborn as sns

# Определение функции для построения графика
def plot_series(series, series_name):
    xs = series['date']
    ys = series['people_fully_vaccinated']
    plt.plot(xs, ys, label=series_name)

# Создание фигуры и осей для графика
fig, ax = plt.subplots(figsize=(10, 5.2))

# Сортировка данных по дате
df_china_sorted = df_china.sort_values('date', ascending=True)

# Построение графика
plot_series(df_china_sorted, 'China')

# Добавление меток к осям
plt.xlabel('Date')
plt.ylabel('People Fully Vaccinated')

plt.show()  # Показать график
```

График новых смертей от COVID-19 в Китае


```python
from matplotlib import pyplot as plt

# Определение функции для построения графика
def plot_series(series, series_name):
    xs = series['date']
    ys = series['new_deaths']
    plt.plot(xs, ys, label=series_name)

# Создание фигуры и осей для графика
fig, ax = plt.subplots(figsize=(10, 5.2))

# Сортировка данных по дате
df_china_sorted = df_china.sort_values('date', ascending=True)

# Построение графика
plot_series(df_china_sorted, 'China')

# Удаление лишних границ графика
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Добавление меток к осям
plt.xlabel('Date')
plt.ylabel('New Deaths')

plt.show()  # Показать график
```

Строим графики по следующим данным:

*   вакцинированных
*   смертей от Ковид-19
*   новые случаи заболеваний Ковид-19






```python
df_china['total_deathsX'] = df_china.total_deaths * 500    #добавляем столбец total_deathsX, значения которого умножаются на 100 от оригинального столбца для лучшего масштабирования на графике
df_china['new_casesX'] = df_china['new_cases'] * 1000 # аналогично создаем новый столбец с масштабированными значениями
df_china.plot (x = 'date', y = [ 'people_vaccinated', 'total_deathsX', 'new_casesX'], figsize = (20, 8)) # рисуем график, используя метод plot датафрейма в pandas
```

сравнительная статистика Китая с другими странами

```python
import matplotlib.pyplot as plt

df1 = df.loc[df['location'] == 'United States',['date', 'total_deaths']]
df2 = df.loc[df['location'] == 'Russia',['date', 'total_deaths']]
df3 = df.loc[df['location'] == 'China', ['date', 'total_deaths']]

plt.figure (figsize = (20, 8)) # зададим размер полотна графика в дюймах
plt.plot (df1['date'], df1['total_deaths'], label = 'USA total deaths')
plt.plot (df2['date'], df2['total_deaths'], label = 'Russia total deaths')
plt.plot (df3['date'], df3['total_deaths'], label = 'China total deaths')
plt.legend(fontsize=20)

plt.show ()
```
